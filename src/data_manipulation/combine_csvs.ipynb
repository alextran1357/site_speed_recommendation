```python
from pathlib import Path
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
VOLUME_PATH = Path("/Volumes/workspace/site-speed-recommendation/psi_data")
onlyfiles = [f for f in listdir(VOLUME_PATH) if isfile(join(VOLUME_PATH, f))]
onlyfiles
```

```python
full_files = []
encoding_log = []
row_count = []

for filename in onlyfiles:
    try:
        df = pd.read_csv(VOLUME_PATH/filename, encoding="utf-8", low_memory=False)
        encoding_log.append((filename, "utf-8"))
    except UnicodeDecodeError:
        df = pd.read_csv(filename, encoding="cp1252", low_memory=False)
        encoding_log.append((filename, "cp1252"))
        
    row_count.append(len(df))
    full_files.append(df)
listing_combined = pd.concat(full_files, axis=0, ignore_index=True)
```
